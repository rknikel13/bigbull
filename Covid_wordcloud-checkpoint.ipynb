```python
import pandas as pd
import lxml.html
import requests
```

```python
res = requests.get('https://korean.cdc.gov/coronavirus/2019-ncov/cases-updates/cases-in-us.html')
res
```

```python
root = lxml.html.fromstring(res.text)
root
```

```python
content = root.cssselect('[class="container d-flex flex-wrap body-wrapper bg-white"]')[0].text_content()
content
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
tdm = cv.fit_transform([content]).toarray()
tdm
```

```python
df = pd.DataFrame({
    'word': cv.get_feature_names(),
    'freq': tdm.flat
})
df
```

```python
df.sort_values('freq', ascending=False).head(20)
```

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(font_path='C:\Windows\Fonts\malgun.ttf', width=400, height=400, background_color='white')
```

```python
frequencies = {word: freq for word, freq in zip(cv.get_feature_names(), tdm.flat)}
```

```python
wc.fit_words(frequencies)
```
